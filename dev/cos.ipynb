```python
import matplotlib.pyplot as plt
from world_machine_experiments.shared.save_metrics import load_metrics, load_multiple_metrics, get_values
import numpy as np
```

```python
metrics = load_multiple_metrics("cosin_test", "toy1d_metrics")
train_history = load_multiple_metrics("cosin_test", "toy1d_train_history")
```

```python
normal = get_values(metrics, ["means", "normal", "state_decoded_mse"])
pred_shallow = get_values(metrics, ["means", "prediction_shallow", "state_decoded_mse"])
pred_local = get_values(metrics, ["means", "prediction_local", "state_decoded_mse"])


fig, ax = plt.subplots(dpi=300, figsize=(35, 8))
plt.bar(list(metrics.keys())[:-1], normal[:-1])
plt.title("normal")
plt.show()

fig, ax = plt.subplots(dpi=300, figsize=(35, 8))
plt.bar(metrics.keys(),pred_shallow)
plt.title("pred_shallow")
plt.show()

fig, ax = plt.subplots(dpi=300, figsize=(35, 8))
plt.bar(list(metrics.keys())[:-1],pred_local[:-1])
plt.title("pred_local")
plt.show()
```
